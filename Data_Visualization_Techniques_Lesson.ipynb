```python
import matplotlib.pyplot as plt
import numpy as np

Grades={'11th', '12th', '9','10'}
StudentsPerGrade={4,52,8,3}

EvenDistance=np.arange(len(Grades))

plt.bar(EvenDistance, StudentsPerGrade, width=0.75, color=['orange','pink','blue','red'
                                                        ], edgecolor='black')
plt.xlabel("grades")
plt.ylabel("num of students")
plt.title("grade distribution")
plt.show()
```

```python
Games = ["1","2","3", "4", "5"]
Warriors = [108, 107, 100, 107, 104]
Celtics = [120, 88, 116, 97, 94]

distance=np.arange(len(Games))
plt.bar(distance-0.2, Warriors, 0.4, label="warriors", color='blue')
plt.bar(distance+0.2, Celtics, 0.4, label="celtics",color='green')

plt.xticks(distance,Games)
plt.xlabel("Game num")
plt.ylabel("points score")
```

```python
plt.plot(Games, Warriors)
plt.plot(Games, Celtics)

plt.title("nba final pts")
plt.show()

plt.plot(Games,Warriors,color='blue',marker='o',label="warriors")

plt.plot(Games, Celtics, color='green', marker='>', label="Celtics")
plt.grid(True)
plt.legend()
plt.xlabel("Game #", fontsize=14)
plt.ylabel("points score",fontsize=14)

plt.title("nba final pts", fontsize=20)
```

```python
astro_labels = ['Aries', 'Taurus', 'Gemini', 'Cancer', 'Leo', 'Virgo', 'Libra', 'Scorpio', 'Sagittarius', 'Capricorn', 'Aquarius', 'Pisces']
students_astro = [7, 4, 5, 6, 6, 5, 6, 4, 10, 4, 5, 5]
students_oneDirection = [3, 3, 2, 2, 1, 2, 4, 1, 3, 1, 2, 3]
d=np.arange(len(astro_labels))
'''
fig=plt.figure()
ax=fig.add_subplot()
'''
fig, ax= plt.subplots()

ax.plot(astro_labels, students_oneDirection)

ax.plot(astro_labels, students_oneDirection, color='black')
ax.set_ylabel("students who know 1d")
ax.set_ylim(0,5)
ax.legend(["Knows 1D"], loc="upper left")
ax.set_xticklabels(astro_labels, rotation = 45, fontsize = 8)


ax2=ax.twinx()
ax2.bar(d, students_astro, width=0.5, color='orange',alpha=0.5)
ax2.set_ylim(0,12)
ax2.legend(["num students"], loc="upper right")
ax2.set_ylabel("students per astro sign")
ax2.grid(False)
plt.show()


fig, ax=plt.subplots()
ax.pie(students_astro, labels=astro_labels, colors=['pink','blue'], startangle=90, autopct='%1.1f%%')
plt.show()
```
